```python
%matplotlib notebook
import numpy as np
import os
import sys
from matplotlib import pyplot as plt
#from matplotlib import animation
#Path to the scripts shared troughout phases
parent_path=os.path.abspath('..')
path_common =  '/'.join(parent_path.split('/')[:-1])
save_path = os.path.join(parent_path,'Moving_Rigidbody','Plots','convergence','no_gravity_fixed_box_fixed_not_in_error')
if(not os.path.exists(save_path)):
    os.makedirs(save_path)
```

```python
basis_path = os.path.join(parent_path,'Moving_Rigidbody','small','no_gravity_fixed_box_fixed_not_in_error')
one_basis = basis_path
logged_iterations = []
loss_tr = []
loss_te = []
loss_ge = []
loss_step_tr = []
loss_step_te = []
loss_step_ge = []
for folder in os.listdir(one_basis):
    path_to_saves = os.path.join(one_basis,folder,'Saves')
    logged_iterations.append(np.load(os.path.join(path_to_saves,'logged_iterations.npy')))
    loss_tr.append(np.load(os.path.join(path_to_saves,'losses_tr.npy')))
    loss_te.append(np.load(os.path.join(path_to_saves,'losses_te.npy')))
    loss_ge.append(np.load(os.path.join(path_to_saves,'losses_ge.npy')))
    loss_step_tr.append(np.load(os.path.join(path_to_saves,'losses_step_tr.npy')))
    loss_step_te.append(np.load(os.path.join(path_to_saves,'losses_step_te.npy')))
    loss_step_ge.append(np.load(os.path.join(path_to_saves,'losses_step_ge.npy')))
    
logged_iterations = np.array(logged_iterations[0])
loss_tr = np.array(loss_tr)
loss_te = np.array(loss_te)
loss_ge = np.array(loss_ge)
loss_step_te = np.array(loss_step_te)
loss_step_ge = np.array(loss_step_ge)
```

```python
loss_tr_mean = np.mean(loss_tr,axis = 0)
loss_tr_std = np.std(loss_tr,axis = 0)
loss_te_mean = np.mean(loss_te,axis = 0)
loss_te_std = np.std(loss_te,axis = 0)
loss_ge_mean = np.mean(loss_ge,axis = 0)
loss_ge_std = np.std(loss_ge,axis = 0)
loss_step_tr_mean = np.mean(loss_step_tr,axis = 0)
loss_step_tr_std = np.std(loss_step_tr,axis = 0)
loss_step_te_mean = np.mean(loss_step_te,axis = 0)
loss_step_te_std = np.std(loss_step_te,axis = 0)
loss_step_ge_mean = np.mean(loss_step_ge,axis = 0)
loss_step_ge_std = np.std(loss_step_ge,axis = 0)


x1 = logged_iterations
y1 = np.log(loss_tr_mean)
e1 = np.log(loss_tr_std)

plt.close('all')
fig = plt.figure(1, figsize=(18, 6))
fig.clf()

ax1 = fig.add_subplot(131)
ax1.plot(x1, y1, "b")
plt.fill_between(x1, y1-0.434*e1/y1, y1+0.434*e1/y1, color = "b",alpha=0.4)

ax1.set_title("Training Error")
ax1.set_xlabel("Epoch number")
ax1.set_ylabel("Rollout Loss (Log Scale)")
#ax1.set_yscale("log")



x2 = logged_iterations
y2 = np.log(loss_te_mean)
e2 = np.log(loss_te_std)
ax2 = fig.add_subplot(132,sharey = ax1)
ax2.plot(x2, y2, "b")
plt.fill_between(x2, y2-0.434*e2/y2, y2+0.434*e2/y2, color = "b",alpha=0.4)

ax2.set_title("Testing Error")
ax2.set_xlabel("Epoch number")
ax2.set_ylabel("Rollout Loss (Log Scale)")
#ax2.set_yscale("log")


x3 = logged_iterations
y3 = np.log(loss_ge_mean)
e3 = np.log(loss_ge_std)

ax3 = fig.add_subplot(133,sharey = ax1)
ax3.plot(x3, y3, "b")
plt.fill_between(x3, y3-0.434*e3/y3, y3+0.434*e3/y3, color = "b",alpha=0.4)

ax3.set_title("Generalisation Error")
ax3.set_xlabel("Epoch number")
ax3.set_ylabel("Rollout Loss (Log Scale)")
#ax3.set_ylabel("Rollout Loss (Log Scale)")





fig.suptitle("Convergence of training, Loss sampled every 250 epoch", fontsize=16)

plt.show()
#Save figure                  
fig.savefig(os.path.join(save_path,'Convergence_small-fixed-box_no-gravity_Rollout.pdf'))
```
