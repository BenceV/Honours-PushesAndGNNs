```python
def combine_list(dict_avg,indr,valr,n):
    
    for e, v in zip(indr,valr):
        es = [e+i for i in range(-250,250)]
        is_not_in = True
        is_in = 0
        
        for ex in es:
            if ex in dict_avg.keys():
                is_not_in = False
                is_in = ex
        
        if is_not_in:
            dict_avg[e] = v
        else:
            dict_avg[is_in] = dict_avg[is_in] + (v - dict_avg[is_in])/n
            
    return dict_avg


def do_help(dict_avg,l,v):
    dict_avg[l]=v 
    return dict_avg
```

```python
dict_avgs_one_tr = dict(zip(one_way_logged_1, one_way_loss_tr_1))
dict_avgs_two_tr = dict(zip(two_way_logged_1, two_way_loss_tr_1))


#dict_avgs_one_tr = combine_list(dict_avgs_one_tr,one_way_logged_1,one_way_loss_tr_1,1)
dict_avgs_one_tr = combine_list(dict_avgs_one_tr,one_way_logged_2,one_way_loss_tr_2,2)
dict_avgs_one_tr = combine_list(dict_avgs_one_tr,one_way_logged_3,one_way_loss_tr_3,3)
dict_avgs_one_tr = combine_list(dict_avgs_one_tr,one_way_logged_4,one_way_loss_tr_4,4)
dict_avgs_one_tr = combine_list(dict_avgs_one_tr,one_way_logged_5,one_way_loss_tr_5,5)

#dict_avgs_two_tr = combine_list(dict_avgs_two_tr,two_way_logged_1,two_way_loss_tr_1,1)
dict_avgs_two_tr = combine_list(dict_avgs_two_tr,two_way_logged_2,two_way_loss_tr_2,2)
dict_avgs_two_tr = combine_list(dict_avgs_two_tr,two_way_logged_3,two_way_loss_tr_3,3)
dict_avgs_two_tr = combine_list(dict_avgs_two_tr,two_way_logged_4,two_way_loss_tr_4,4)
dict_avgs_two_tr = combine_list(dict_avgs_two_tr,two_way_logged_5,two_way_loss_tr_5,5)
```

```python
plt.close('all')
fig = plt.figure(1, figsize=(18, 6))
fig.clf()

ax = fig.add_subplot(1, 1, 1)
ks = list(dict_avgs_two_tr.keys())
vs = list(dict_avgs_two_tr.values())
ax.plot(ks, np.log(vs), "r",label="Two way edge")
ks = list(dict_avgs_one_tr.keys())
vs = list(dict_avgs_one_tr.values())
ax.plot(ks, np.log(vs), "b",label="One way edge")
ax.legend()
ax.set_title("Next step loss: Training")
ax.set_xlabel("")
plt.show()

#Save figure                  
fig.savefig(os.path.join(save_path,"ConvergencePlots_avg_tr_log_scale.png"))
```
